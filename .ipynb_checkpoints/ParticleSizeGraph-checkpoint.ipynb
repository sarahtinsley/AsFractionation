# Read data

```python
filepath_PSA = r'C:\Users\sbtin3\Documents\Python\Doc\ParticleSizeCleaned.csv'
df_PSA = pd.read_csv (filepath_PSA)
```

# Define Data

```python
MR = df_PSA.head(4)
CG = df_PSA.iloc[[4,5,6,7]]
JF = df_PSA.iloc[[8,9,10,11]]
LG = df_PSA.iloc[[12,13,14,15]]
CN = df_PSA.iloc[[16,17,18,19]]
```

```python
#I want to make an array of numbers that fit each case
#1. Average size distribution of each site
#MR
MRS1mean = np.mean(MR['S1'])
MRS2mean = np.mean(MR['S2'])
MRS3mean = np.mean(MR['S3'])
MRS4mean = np.mean(MR['S4'])
MRS5mean = np.mean(MR['S5'])
MRS6mean = np.mean(MR['S6'])

#CG
CGS1mean = np.mean(CG['S1'])
CGS2mean = np.mean(CG['S2'])
CGS3mean = np.mean(CG['S3'])
CGS4mean = np.mean(CG['S4'])
CGS5mean = np.mean(CG['S5'])
CGS6mean = np.mean(CG['S6'])

#JF
JFS1mean = np.mean(JF['S1'])
JFS2mean = np.mean(JF['S2'])
JFS3mean = np.mean(JF['S3'])
JFS4mean = np.mean(JF['S4'])
JFS5mean = np.mean(JF['S5'])
JFS6mean = np.mean(JF['S6'])

#LG
LGS1mean = np.mean(LG['S1'])
LGS2mean = np.mean(LG['S2'])
LGS3mean = np.mean(LG['S3'])
LGS4mean = np.mean(LG['S4'])
LGS5mean = np.mean(LG['S5'])
LGS6mean = np.mean(LG['S6'])

#CN
CNS1mean = np.mean(CN['S1'])
CNS2mean = np.mean(CN['S2'])
CNS3mean = np.mean(CN['S3'])
CNS4mean = np.mean(CN['S4'])
CNS5mean = np.mean(CN['S5'])
CNS6mean = np.mean(CN['S6'])
```

# Plots

```python
#This plot will show the different size fraction averages for all depths, separated by site
labels = ['MR', 'CG', 'JF', 'LG', 'CN']

#this defines my categories that go up the columns (the fractions), using the means I made in the previous box. Turning the values into arrays let me stack them properly later on.
S1= np.array([MRS1mean,CGS1mean,JFS1mean,LGS1mean,CNS1mean])
S2= np.array([MRS2mean,CGS2mean,JFS2mean,LGS2mean,CNS2mean])
S3= np.array([MRS3mean,CGS3mean,JFS3mean,LGS3mean,CNS3mean])
S4= np.array([MRS4mean,CGS4mean,JFS4mean,LGS4mean,CNS4mean])
S5= np.array([MRS5mean,CGS5mean,JFS5mean,LGS5mean,CNS5mean])
S6= np.array([MRS6mean,CGS6mean,JFS6mean,LGS6mean,CNS6mean])

#this is the width of the bars/columns.
width = 0.5       

#BEGINNING OF ACTUAL PLOT 
fig, ax = plt.subplots(figsize=(7,7))
plt.bar(labels, S1, width, label='0-2um', color='hotpink', hatch = "/")
plt.bar(labels, S2, width, bottom=S1,label='2-62um',color='orange', hatch = "o")
plt.bar(labels, S3, width, bottom=S1+S2,label='63-212um',color='gold', hatch = "x")
plt.bar(labels, S4, width, bottom=S1+S2+S3,label='212-600um',color='limegreen', hatch = "-")
plt.bar(labels, S5, width, bottom=S1+S2+S3+S4,label='600-1200um',color='lightskyblue')
plt.bar(labels, S6, width, bottom=S1+S2+S3+S4+S5,label='1200um+',color='mediumpurple', hatch = ".")

#Titles and Labels
ax.set_title('Particle Size by Site', fontsize=20)
ax.set_ylabel('Particle Size as % of Total',fontsize=20)
ax.set_xlabel('Site',fontsize=20)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=15) #this is the site names/percentages

plt.show()
```
